## 1. Постановка задачи и описание алгоритма

В связи с учащением проведения повторных разработок месторождений возникает ситуация, в которой обработчикам геофизической информации приходиться иметь дело с геологическими съемками одной геологической среды, но проведенных в разное время и новой аппаратурой. Ясно, что полезная составляющая сигнала в обоих случаях одинаковая, однако реализация шума и частотные характеристиги сигнала будут совершенно разные. Поэтому использование информации с обеих съемок предоставляет возможность точнее оценивать поведение сигнала на разных частотах. Что в дальнейшем сулит более устойчивой моделью геологической среды.

Описываемая мной программа предназначена для комбинирования сейсимческой информации, полученной при разных съемках одной местности, оптимальным образом. В качестве критерия оптимальности было выбрано отношение сигнал/шум на каждой частоте.

Идея алгоритма заключается в следующем:
* Отсортировать сигналы по огт. То есть получить набор сейсмограмм N_{ji}, соответсвующих возбуждению источника на одинковых пикетах.
* Последовательно из каждой пары сейсмотрасс (i, i+1) i - any каждой сейсограммы j выделяется плавующее окно.
* В окне считается взаимно-корелляционная, средняя авто-корреляциянная функции и по формуле (вставить) определяется отношение сигнал/шум. После окно сдвигается и алгоритм повторяется, пока не будет достигнут конец трассы
* В результате для каждой трассы получается набор весовых коэффциентов, который остается покоординатно отнормировать на сумму соотвествующих коэффициентов оставшихся трасс с таким же позиционным номером
* Умножив все трассы на полученный набор взвещенных весовых коэффициентов и проссмуировав сейсмотрассы, получаем оптимальную сумму соотвествующих сигналов (трасс).
